```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample data (you can replace this with your own dataset)
data = {
    'Name': ['John', 'Jane', 'Mike', 'Emily', 'Alex'],
    'Age': [25, 30, 22, 28, 35],
    'Score': [90, 85, 78, 95, 88]
}

# Create a DataFrame from the data
df = pd.DataFrame(data)

# Display the DataFrame
print("DataFrame:")
print(df)

# Data Analysis
print("\nData Analysis:")
print("Mean Age:", df['Age'].mean())
print("Maximum Score:", df['Score'].max())
print("Minimum Score:", df['Score'].min())

# Data Visualization
plt.figure(figsize=(8, 4))
plt.bar(df['Name'], df['Score'], color='skyblue')
plt.xlabel('Name')
plt.ylabel('Score')
plt.title('Scores of Students')
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Sample email data
data = {
    'sender': ['alice@example.com', 'bob@example.com', 'alice@example.com'],
    'receiver': ['bob@example.com', 'alice@example.com', 'carol@example.com'],
    'subject': ['Hello', 'Meeting Reminder', 'Project Update'],
    'timestamp': ['2023-08-01 10:00:00', '2023-08-02 14:30:00', '2023-08-03 09:15:00'],
    'content': [
        'Hi Bob,\n\nHow are you?',
        'Hi Alice,\n\nDon\'t forget the meeting at 3 PM.',
        'Hi Carol,\n\nHere\'s the latest project update.'
    ]
}

# Create DataFrame
df = pd.DataFrame(data)

# Convert timestamp to datetime
df['timestamp'] = pd.to_datetime(df['timestamp'])

# Save to CSV
df.to_csv('emails.csv', index=False)
print("CSV file created successfully.")

# Load from CSV
df = pd.read_csv("emails.csv")
df['timestamp'] = pd.to_datetime(df['timestamp'])

# Drop missing values if any
df.dropna(inplace=True)

# Add a new column for email length
df['email_length'] = df['content'].apply(len)

# 📊 Histogram of Email Lengths
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='email_length', bins=30, kde=True)
plt.xlabel('Email Length')
plt.ylabel('Count')
plt.title('Distribution of Email Lengths')
plt.show()

# 📊 Top 10 Email Senders
top_senders = df['sender'].value_counts()[:10]
plt.figure(figsize=(12, 6))
sns.barplot(x=top_senders.index, y=top_senders.values)
plt.xticks(rotation=45)
plt.xlabel('Sender')
plt.ylabel('Number of Emails')
plt.title('Top 10 Email Senders')
plt.tight_layout()
plt.show()

# 📈 Email Activity Over Time
df['year_month'] = df['timestamp'].dt.to_period('M')
email_activity = df.groupby('year_month').size()

plt.figure(figsize=(12, 6))
email_activity.plot(kind='line', marker='o')
plt.xlabel('Year-Month')
plt.ylabel('Number of Emails')
plt.title('Email Activity Over Time')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import pandas as pd

# 1. Fixing the array for proper DataFrame creation
data = np.array([['Col1', 'Col2', 'Col3'], ['Row1', 1, 2], ['Row2', 3, 4]])
df1 = pd.DataFrame(data=data[1:, 1:], index=data[1:, 0], columns=data[0, 1:])
print("1. DataFrame from 2D array:")
print(df1)

# 2. 2D Array to DataFrame
my_2darray = np.array([[1, 2, 3], [4, 5, 6]])
print("\n2. DataFrame from 2D NumPy array:")
print(pd.DataFrame(my_2darray))

# 3. Dictionary to DataFrame
my_dict = {1: ['1', '3'], 2: ['1', '2'], 3: ['2', '4']}
print("\n3. DataFrame from Dictionary:")
print(pd.DataFrame(my_dict))

# 4. Creating a simple DataFrame
my_df = pd.DataFrame(data=[4, 5, 6, 7], index=range(0, 4), columns=['A'])
print("\n4. Custom DataFrame:")
print(my_df)

# 5. Series to DataFrame (country-capital)
my_series = pd.Series({
    "United Kingdom": "London",
    "India": "New Delhi",
    "United States": "Washington",
    "Belgium": "Brussels"
})
print("\n5. DataFrame from Series:")
print(pd.DataFrame(my_series, columns=["Capital"]))

# 6. Shape and length of DataFrame
df = pd.DataFrame(np.array([[1, 2, 3], [4, 5, 6]]))
print("\n6. Shape and number of rows of DataFrame:")
print("Shape:", df.shape)
print("Number of rows (index length):", len(df.index))
```
